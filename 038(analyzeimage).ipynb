```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
blank_w = np.ones((500,500,3),np.uint8) * 255
blank_b = np.zeros((500,500,3),np.uint8) * 255

# hist = cv2.calcHist([blank_w],[0],None,[256],[0,256])
# plt.plot(hist)
# plt.show()

hist = cv2.calcHist([blank_b],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()

# cv2.imshow('blank w',blank_w)
cv2.imshow('blank b',blank_b)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img1 = cv2.imread('A01.jfif')
img2 = cv2.imread('A02.jfif')
img3 = cv2.imread('A03.jfif')

hist1 = cv2.calcHist([img1],[0],None,[256],[0,256])
hist2 = cv2.calcHist([img2],[0],None,[256],[0,256])
hist3 = cv2.calcHist([img3],[0],None,[256],[0,256])
plt.plot(hist1)
plt.plot(hist2)
plt.plot(hist3)
plt.show()

cv2.imshow('img1',img1)
cv2.imshow('img2',img2)
cv2.imshow('img3',img3)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img1 = cv2.imread('A01.jfif')

b ,g, r = cv2.split(img1)

plt.hist(b.ravel(),255,(0,255),color='blue')
plt.hist(g.ravel(),255,(0,255),color='green')
plt.hist(r.ravel(),255,(0,255),color='red')

plt.show()

cv2.imshow('img1',img1)
cv2.imshow('img2',b)
cv2.imshow('img3',g)
cv2.imshow('img4',r)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
