```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('A01.jfif')
img = cv2.resize(img,(500,500))

gry = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

hist = cv2.calcHist([gry],[0],None,[255],[0,255])
eq = cv2.equalizeHist(gry)
hist1 = cv2.calcHist([eq],[0],None,[255],[0,255])


plt.plot(hist)
plt.plot(hist1,c = 'red')
plt.show()
cv2.imshow('img1',eq)
cv2.imshow('img',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('A02.jfif')
img = cv2.resize(img,(500,500))

gry = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

hist = cv2.calcHist([gry],[0],None,[255],[0,255])
eq = cv2.equalizeHist(gry)
hist1 = cv2.calcHist([eq],[0],None,[255],[0,255])


cl = cv2.createCLAHE(clipLimit=1,tileGridSize=(8,8))
f = cl.apply(gry)


plt.plot(hist)
plt.plot(hist1,c = 'red')
plt.plot(f,c = 'green')
plt.show()

cv2.imshow('wse',f)
cv2.imshow('img1',eq)
cv2.imshow('img',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
